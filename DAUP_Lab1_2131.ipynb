```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/content/insurance.csv")

# Vectors, Factors, Lists, Matrices, and Data Frames

ages = np.array(data['age'])

expenses = np.array(data['expenses'])

smoker_factor = pd.Categorical(data['smoker'], categories=["no", "yes"])

example_list = [ages[:5], smoker_factor[:5], expenses[:5]]

matrix_example = np.array([ages[:10], expenses[:10]])

data_frame_example = data[['age', 'smoker', 'expenses']].head()

print("Vector (ages):", ages[:5])
print("Factor (smoker):", smoker_factor[:5])
print("List:", example_list)
print("Matrix:\n", matrix_example)
print("DataFrame:\n", data_frame_example)
```

```python
# Scatter Plot: Age vs Smoking (by Expenses)
plt.figure(figsize=(8, 6))
for smoker_status, group_data in data.groupby('smoker'):
    plt.scatter(group_data['age'], group_data['expenses'], label=smoker_status, alpha=0.7)
plt.title('Scatter Plot: Age vs Expenses by Smoking Status')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend(title="Smoker")
plt.grid(True)
plt.show()
```

```python
# Scatter Plot: Age (20–40) vs Smoking (by Expenses)
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]
plt.figure(figsize=(8, 6))
for smoker_status, group_data in filtered_data.groupby('smoker'):
    plt.scatter(group_data['age'], group_data['expenses'], label=smoker_status, alpha=0.7)
plt.title('Scatter Plot: Age (20-40) vs Expenses by Smoking Status')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.legend(title="Smoker")
plt.grid(True)
plt.show()
```

```python
# Bar Plot: Average Expenses of Smokers vs Non-Smokers
avg_expenses = data.groupby('smoker')['expenses'].mean()
plt.figure(figsize=(8, 6))
avg_expenses.plot(kind='bar', color=['green', 'red'], edgecolor='black')
plt.title('Average Expenses: Smokers vs Non-Smokers')
plt.xlabel('Smoker')
plt.ylabel('Average Expenses')
plt.xticks(rotation=0)
plt.show()
```

```python
# Bar Plot: Number of Smokers Gender-wise
smoker_gender_count = data[data['smoker'] == 'yes']['sex'].value_counts()
plt.figure(figsize=(8, 6))
smoker_gender_count.plot(kind='bar', color=['blue', 'pink'], edgecolor='black')
plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Scatter Plot: Age vs Expenses
plt.figure(figsize=(8, 6))
plt.scatter(data['age'], data['expenses'], c='green')
plt.title('Scatter Plot: Age vs Expenses')
plt.xlabel('Age')
plt.ylabel('Expenses')
plt.grid(True)
plt.show()
```

```python
# Counting the number of teenage smokers (ages 13 to 19)
teenage_smokers_count = data[(data['age'] >= 13) & (data['age'] <= 19) & (data['smoker'] == 'yes')].shape[0]
print(f"Number of teenage smokers: {teenage_smokers_count}")
```
